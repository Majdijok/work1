```python
count = df_can['PdDistrict'].value_counts().tolist()
print(count)
```

```python
count1 = df_can.groupby(['PdDistrict']).count()
count1.head(12)
```

```python
count1.drop(['Descript', 'DayOfWeek', 'Date', 'Time'], axis=1, inplace=True)
```

```python
count1.drop(['Resolution', 'Address', 'X', 'Y','Location','PdId'], axis=1, inplace=True)
count1.head(12)
```

```python
count1.rename(columns={'Category':'Count','PdDistrict':'Neighbourhood'}, inplace=True)
count1.head(12)
```

```python
count1=count1.reset_index()
count1.rename(columns={'Category':'Count','PdDistrict':'Neighbourhood'}, inplace=True)

count1
```
