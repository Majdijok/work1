```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

# First Task

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
```

```python

df
```

# Second Task


```python
df.drop(df[df.Borough == 'Not assigned'].index, inplace=True)
df.reset_index(drop=True, inplace=True)
#Removal of borough that is Not assigned
```

```python
df['Neighbourhood'].replace('Not assigned', df.Borough, inplace=True)
#Change of Not assigned Neighbour as the same as their borough name
```

```python
df.shape
```

```python
#Import the geographical coordinates of each postal code
df1 = pd.read_csv ('http://cocl.us/Geospatial_data')
df1.head()
```

```python
df1.sort_values(by=['Postal Code'], inplace=True)
df.sort_values(by=['Postal Code'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
result = pd.concat([df, df1['Latitude'],df1['Longitude']], axis=1, sort=False)
```

```python
result.head()
```

```python
result.loc[result['Borough'].str.contains('Toronto'),'TorontoCity']='True' 
#Retreive Toronto related Borough
```

```python
result
```

```python
result.dropna(inplace=True)
del result['TorontoCity']
result.reset_index(drop=True, inplace=True)
#Eliminate all the NaN from the Toronto related new table
```

```python
neighborhoods=result
neighborhoods
```

```python
address = 'Toronto, TO'

geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```
